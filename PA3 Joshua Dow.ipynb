```python
a = int(input('Enter the length of side A: '))
b = int(input('Enter the length of side B: '))
c = int(input('Enter the length of side c: '))
def calcPerimeter(a,b,c):
    calcPerimeter = a + b + c
    return calcPerimeter
print( 'Perimeter is: ', calcPerimeter(a,b,c))
def calcArea(a,b,c):
    s = (a + b + c) / 2
    calcArea = (s*(s-a)*(s-b)*(s-c)) ** 0.5
    return calcArea
print('The Area is: ', calcArea(a,b,c))
```

```python
import random
number = random.randint(1,1000)
win = False
while win==False:
    guess = input("Enter Your Guess: ")
    if number==int(guess):
        print("Correct!")
        win == True
        break
    elif number>int(guess):
        print("Too Low! Guess Again")
    else:
        print("Too High! Guess Again")
```
